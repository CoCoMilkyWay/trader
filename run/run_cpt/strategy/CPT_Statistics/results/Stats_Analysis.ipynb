- First we determine the trailing stops and holding period we want to participate:
  - we see that most trailing stops method alone can make you money without hefty trading fees, but with fees, we have to hold longer and trade less-frequently
  - except some special methods, most have similar performance
  - as hold time increases, profit standard deviation increases, which potentially give you larger win and losses
  - your hands are pretty much tied here, there are not a lot you can do to improve trailing stops when it comes to TP/SL in high-margin/volatility markets with somewhat good market efficiency, or without lots of event-driven market impacts(thus technical analysis prevails)
  - most profit/hold distributions lean to Gaussian, some performant methods leans toward fat-tailed student-t, especially when holds are longer(e.g. >2hours)
  - median are far away from mean in the lossing side, which means these method would mostly give small losing trades(open/close/funding fees considered), but there are potentially quite a few trades that yield considerable profits
  - it is the strategy's job to filter out losing trade(both large and small)
  - better not participate trading under 1 hour, as max profit potential is greatly limited and we are facing a negative mean(because of fees)
  - performance shift(decay) over different years are not significant, because we are analyzing short hold periods(<10hours). However, there is no doubt that this timeframe is very competitive (players competing to harvest market liquidity)

- Second because we are trading reversal scalping, and we know the average holding period, we need to find an average that shows us the current deviation
  - the MA we choose is VWMA, Why?
    - we dont trade golden/death cross here, we dont need zero-lag to open and close our position
    - we only use this MA as measure of elasticity(thus reverse)
    - we need it to include volume to better represent short-term market true average price
    - cannot use VWAP because this is 7*24 exchange and VWAP deviates from price over time
  - only 1 thing to consider, the period
    - because we have determined the holding period we want to participate in, we know the expected cross time. From this, we can determine the expected period of MA that we want to use as long/short reversal(best pnl) reference
    - better match both mean and std deviation of target holding period and MA crossover period
    - are there ways to twink it better? yes, but this is a good start, align the reversal periods first
    - because we are looking for 5 hrs hold period(mean=5hrs, median=2hrs, we need to match median, not mean), we want that in all period of holding, majority is marked as oversold or overbought, which according to our research data, matchs with window length N

  - How to define "quality" of volatility at certain period?
    - we did see some peak frequency components at n-hours period using Fourier Analysis. However, the period of volatility anomaly is irregular and drift overtime. Thus Fourier analysis on trading data is not really a good idea to gather insight
    - because of asymmetrical response to long/short, non-linearity of how market price shifts, the better metric to determine "quality" of volatility at certain period is better studies by other metrics like:
      - TP/SL parameter searching: this is the method everyone uses, and good quality trend is simply trend that has large magnitude and not stopping people out, this is naturally the better definition of the goodness of the volatility(trend in small time-frames)
      - for every single point in k bar, what is the max magnitude that the deviation can occur before price goes back to this level? we are expecting to see some very high-peaked distribution here, but if there are any huge/stable peaks at tail region, it provides very good insights to us too
